```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mydata=pd.read_csv(r"D:\DataSets-master\tango\german_credit_data.csv")
```

```python
mydata
```

```python
mydata.keys()
```

```python
mydata1=mydata.drop(["Unnamed: 0"],axis=1)
```

```python
mydata1
```

```python
mydata1['Saving accounts'] = mydata1['Saving accounts'].map({"little":0,"moderate":1,"quite rich":2 ,"rich":3})
```

```python
mydata1['Checking account'] = mydata1['Checking account'].map({"little":0,"moderate":1,"rich":2 });
```

```python
mydata1
```

```python
mydata1['Sex'] = mydata1['Sex'].map({"male":0,"female":1}).astype(float);



mydata1['Housing'] = mydata1['Housing'].map({"own":0,"free":1,"rent":2}).astype(float);





mydata1['Purpose'] = mydata1['Purpose'].map({'radio/TV':0, 'education':1, 'furniture/equipment':2, 'car':3, 'business':4,
       'domestic appliances':5, 'repairs':6, 'vacation/others':7}).astype(float);
```

```python
mydata1
```

```python
plt.scatter(mydata1["Purpose"],mydata1["Credit amount"])
plt.xlabel("Purpose")
plt.ylabel("Credit amount")
plt.show()
```

```python
plt.scatter(mydata1["Age"],mydata1["Credit amount"])
plt.xlabel("Age")
plt.ylabel("Credit amount")
plt.show()
```

```python
plt.scatter(mydata1["Duration"],mydata1["Credit amount"])
plt.xlabel("Duration")
plt.ylabel("Credit amount")
plt.show()
```

```python
plt.scatter(mydata1["Job"],mydata1["Credit amount"])
plt.xlabel("Job")
plt.ylabel("Credit amount")
plt.show()
```
